# **Uso API con Pandas**

Una API es una interfaz de acceso programable, es lo que utilizan muchas aplicaciones para especificar cómo comunicarse. En este caso se va a utilizar la API de datos de *COVID19* que ofrece la información por países sin restricciones a través de la url https://api.covid19api.com/countries

```python
import pandas as pd
```

```python
url = 'https://api.covid19api.com/countries'
```

```python
url
```

```python
df = pd.read_json(url)
```

```python
df
```

```python
df[df['Country'] == 'Spain']
```

# **Tiempo real: España**

```python
url_rt_es = 'https://api.covid19api.com/country/spain/status/confirmed/live'
df_rt_es = pd.read_json(url_rt_es)
df_rt_es
```

```python
df_rt_es.head()
```

```python
df_rt_es.tail()
```

```python
casos_es = df_rt_es.set_index('Date')['Cases']
casos_es.plot(title="Casos de Covid19 en España")
```

# **Ahora, se realizará el proceso con Panamá**

```python
df[df['Country'] == 'Panama']
```

```python
url_rt_pa = 'https://api.covid19api.com/country/panama/status/confirmed/live'
df_rt_pa = pd.read_json(url_rt_pa)
```

```python
print(df_rt_pa)
```

```python
df_rt_pa
```

```python
df_rt_pa.head()
```

```python
df_rt_pa.tail()
```

```python
casos_pa = df_rt_pa.set_index('Date')['Cases']
casos_pa.plot(title="Casos de Covid19 en Panamá")
```

# **Gráfica de dos países**

```python
casos_pa
```

```python
casos_es
```

```python
pa_vs_es= pd.concat([casos_es, casos_pa], axis=1)
pa_vs_es
```

```python
pa_vs_es.columns = ('España', 'Panamá')
pa_vs_es
```

```python
pa_vs_es.plot(title= "Comparativa Covid19 España-Panama")
```
